```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
train_data=pd.read_csv("Questionnaire.csv", sep=";")
train_data.head()
```

```python
train_data.drop(['Horodateur'],axis = 'columns',inplace=True)

inputs=train_data.apply(LabelEncoder().fit_transform)

inputs
```

```python
# seperate the independent and target variable on training data
target = inputs['Domaine après Baccalauréat']
inputs.drop(columns=['Domaine après Baccalauréat'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(inputs,target,test_size=0.4)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
# fit the model with the training data
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
Y_test[0:10]
```

```python
#la prediction de Y_test à partie de X_test
resultat=model.predict(X_test[0:10])
resultat
```
